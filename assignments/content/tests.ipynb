```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.stats import entropy

def image_entropy(image):
    # Convert image to grayscale

    # Calculate histogram
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])

    # Normalize histogram
    hist_norm = hist.ravel() / hist.sum()

    # Compute entropy
    entropy_value = entropy(hist_norm, base=2)

    return entropy_value

def image_entropy2(image):
    # Convert image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Calculate histogram
    hist = cv2.calcHist([gray_image], [0], None, [256], [0,256])
    
    # Normalize histogram
    hist_norm = hist.ravel() / hist.sum()
    
    # Compute entropy
    entropy_value = entropy(hist_norm, base=2)
    
    return entropy_value
```

```python
# from skimage.restoration import denoise_nl_means


image=cv2.imread('dataset/models/ref18.png')[:,:,::-1]
print('ref18.png',image_entropy2(image))
plt.figure(figsize=(8,5))
plt.subplot(131)
plt.title(f'{image_entropy(image[:,:,0])}')
plt.imshow(image[:,:,0],cmap='gray')
# plt.figure()

plt.subplot(132)
plt.title(f'{image_entropy(image[:,:,1])}')
plt.imshow(image[:,:,1],cmap='gray')
# plt.figure()

plt.subplot(133)
plt.title(f'{image_entropy(image[:,:,2])}')
plt.imshow(image[:,:,2],cmap='gray')
plt.show()

image=cv2.imread('dataset/models/ref26.png')[:,:,::-1]
print('ref26.png',image_entropy2(image))
plt.figure(figsize=(8,5))
plt.subplot(131)
plt.title(f'{image_entropy(image[:,:,0])}')
plt.imshow(image[:,:,0],cmap='gray')
# plt.figure()

plt.subplot(132)
plt.title(f'{image_entropy(image[:,:,1])}')
plt.imshow(image[:,:,1],cmap='gray')
# plt.figure()

plt.subplot(133)
plt.title(f'{image_entropy(image[:,:,2])}')
plt.imshow(image[:,:,2],cmap='gray')
plt.show()

image=cv2.imread('dataset/models/ref27.png')[:,:,::-1]
print('ref27.png',image_entropy2(image))
plt.figure(figsize=(8,5))
plt.subplot(131)
plt.title(f'{image_entropy(image[:,:,0])}')
plt.imshow(image[:,:,0],cmap='gray')
# plt.figure()

plt.subplot(132)
plt.title(f'{image_entropy(image[:,:,1])}')
plt.imshow(image[:,:,1],cmap='gray')
# plt.figure()

plt.subplot(133)
plt.title(f'{image_entropy(image[:,:,2])}')
plt.imshow(image[:,:,2],cmap='gray')
plt.show()
```
